```python
import pandas as pd 
import matplotlib.pyplot as plt
from glob import glob
```

```python
MAXUSERS = []
for idx in range(30):
     df = pd.read_csv(f'gen_{idx}.csv')
     df.drop(columns='Unnamed: 0', inplace=True)
     df.sort_values(by=['Eval'], ascending=False)

     maxUsers = df['Users'].max()
     MAXUSERS.append(maxUsers)
```

```python
plt.figure(figsize=(12,7), dpi=120)
plt.plot(MAXUSERS)
plt.xlabel("Generation")
plt.ylabel("N# Users Satisfied (>360 s)")
```
